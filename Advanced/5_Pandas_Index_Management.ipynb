```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

df.reset_index()

```python
# Filter to United States only, and name the index column
df_usa = df.copy()
df_usa = df_usa[df_usa['job_country'] == 'United States']
df_usa.index.name = 'job_index'
```

```python
# reset_index() to get index by sequential
df_usa.reset_index(inplace=True)
df_usa
```

df.set_index()

```python
# set_index() to set the index by column we specified, this also remove all existing index
df_usa.set_index('job_index', inplace=True)
df_usa
```

df.sort_index()
