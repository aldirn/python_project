```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
# Filter NaN values
df_salary = df[pd.notna(df['salary_year_avg'])].copy()
df_salary['salary_year_avg']
```

```python
# Define a Function
def salary_increase(salary):
    return salary * 1.03

df_salary['salary_year_inflated'] = df_salary['salary_year_avg'].apply(salary_increase)
df_salary[['salary_year_avg', 'salary_year_inflated']]
```

```python
df_salary['salary_year_inflated'] = df_salary['salary_year_avg'].apply(lambda x: x * 1.03)
df_salary[['salary_year_avg', 'salary_year_inflated']]
```

```python
df_salary['salary_year_inflated'] = df_salary['salary_year_avg'] * 1.03
df_salary[['salary_year_avg', 'salary_year_inflated']]
```

Before we calculate by the column, now is for row.
Let's say for Senior title get 5% and the rest is 3%


```python
def projected_salary(row):
    if 'Senior' in row['job_title_short']:
        return 1.05 * row['salary_year_avg']
    else:
        return 1.03 * row['salary_year_avg']

df_salary['salary_year_inflated_v2'] = df_salary.apply(projected_salary, axis=1)
df_salary[['job_title_short', 'salary_year_avg', 'salary_year_inflated_v2']]
```

```python
df_salary['salary_year_inflated_v3'] = df_salary.apply(
    lambda row: 1.05 * row['salary_year_avg'] if 'Senior' in row['job_title_short']
    else 1.03 * row['salary_year_avg'], axis=1)
df_salary[['job_title_short', 'salary_year_avg', 'salary_year_inflated_v2', 'salary_year_inflated_v3']]
```
